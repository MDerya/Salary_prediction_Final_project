```python
#TL OLAN MAAŞLARI DÜZENLEYELİM
num=[]

for item in df["SALARY_FOR_TL_CURRENCY"]:
    if item:
        num.append(str(item).split(" "))
```

```python
#TL tipinden minimum maaş bilgilerini aldık.NaN ifadeler yerine 0 atadık.Değerleri doğru almak için 1000 ile çarptık
num1=[]
for i in range(0,len(num)):
    if num[i][0]=="nan":
        num1.append(0)
    else:
        num1.append(float(num[i][0])*1000)
```

```python
#TL tipinden maximum maaş bilgilerini aldık,
# NaN ifadeler yerine 0 atadık, doğru değerleri elde etmek için 1000 ile çarptık.
# Max ve Min değerleri aynı olmasın diye içerisinde üzeri yazan değerlerin üst sınırını 60000 tl aldık.
num2=[]
for i in range(0,len(num)):
    if num[i][0]=="nan":
        num2.append(0)
    elif (num[i][3])=='üzeri':
        num2.append(60000.0)
    else:
        num2.append(float(num[i][3])*1000)
```

```python
#veri setimize min tl ve max tl kazançlarını gösteren kolonları ekledik.
df["MIN_TL"]=num1

df["MAX_TL"]=num2
```

```python
#TL dışındaki para birimlerini parçaladık.
cur=[]

for item in df["SALARY_FOR_OTHER_CURRENCY"]:
    if item:
        cur.append(str(item).split(" "))
```

```python
#Döviz tipi min maaş bilgilerini çektik.1000 ile çarptım. yanlış girilen 1.0 YERİNE 1000 döviz birimini aynı bıraktık
cur1=[]
for i in range(0,len(cur)):
    if cur[i][0]=="nan":
        cur1.append(0)
    elif cur[i][0]=="1000":
        cur1.append(float(cur[i][0]))
    else:
        cur1.append(float(cur[i][0])*1000)
```

```python
#Döviz tipi max maaş bilgilerini çektik.1000 ile çarptık. Üst sınırı 10999 kabul ettik.(TL de 60000 almıştık.)
cur2=[]
for i in range(0,len(cur)):
    if cur[i][0]=="nan":
        cur2.append(0)
    elif (cur[i][2])=='üzeri':
        cur2.append(10999)#float(cur[i][0])
    else:
        cur2.append(float(cur[i][2])*1000)
```

```python
df["CUR_MIN"]=cur1
df["CUR_MAX"]=cur2
```
